```python
import os
import pandas as pd
import numpy as np
import gensim
import pyLDAvis.gensim_models
import json
from gensim.models import CoherenceModel
```

## Load data

```python
with open("lda_corpus.json", 'r') as f:
    corpus = json.load(f)
```

```python
dictionary = gensim.models.ldamodel.LdaModel.load("id2word")
```

```python
with open("tokenized_bigram_train.json", "r") as f:
    tokenized_bigram_train = json.load(f)
```

## Create Visualization of Topic Models using pyLDAvis

```python
# create list of models
models = []
num_topics = [4, 5]
alpha = [0.05, 0.3, 0.6, "asymmetric", "symmetric"]
eta = [0.05, 0.3, 0.6, "symmetric"]
for topic in num_topics:
    for a in alpha:
        for b in eta:
            models.append("lda_" + str(topic) + "_" + str(a) + "_" + str(b) + ".model")
```

```python
# visualize topic sparsity and topic terms for each model in list
lda_model = gensim.models.ldamodel.LdaModel.load(models[39])
lda_visualization = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_visualization)
```

```python
# evaluate last models based on coherence scores
final_models = ["lda_4_0.05_0.05.model", 'lda_4_0.05_0.3.model', 'lda_4_0.3_0.6.model', 'lda_4_0.6_0.05.model'
               ,'lda_4_asymmetric_0.05.model']
coherence_scores = []
for model in final_models:
    lda_model = gensim.models.ldamodel.LdaModel.load(model)
    coherence = CoherenceModel(lda_model, texts = tokenized_bigram_train, dictionary = dictionary, coherence = "c_v")
    score = coherence.get_coherence()
    coherence_scores.append([model, score])
```

```python
print(pd.DataFrame(coherence_scores))
print("Model with highest coherence score:", coherence_scores[(np.argmax(pd.DataFrame(coherence_scores).loc[:, 1]))][0],
      "Coherence Score:", coherence_scores[(np.argmax(pd.DataFrame(coherence_scores).loc[:, 1]))][1])
## final choice: alpha = 0.05, eta = 0.3
```

```python
# Visualization of final model
final_lda_model = gensim.models.ldamodel.LdaModel.load("lda_4_0.05_0.3.model")
lda_visualization = pyLDAvis.gensim_models.prepare(final_lda_model, corpus, dictionary, sort_topics=False)
pyLDAvis.display(lda_visualization) # choose lambda as 0.15
```
